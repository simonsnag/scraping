### Finding distance between actors
---

```python
actors_dist = {}
actors_name = list(actors.keys())

for i in range(len(actors_name)):
    for j in range(i + 1, len(actors_name)):
        key1 = actors_name[i]
        key2 = actors_name[j]

        actors_dist[(key1, key2)] = get_movie_distance(actors[key1], actors[key2], log_handler=True)
        actors_dist[(key2, key1)] = actors_dist[(key1, key2)]
```

### Creating table and saving it
---

```python
distances = pd.DataFrame(list(actors_dist.items()), columns=['actors', 'distance'])
distances[['start_actor', 'end_actor']] = pd.DataFrame(distances['actors'].tolist(), index=distances.index)
distances.drop(columns=['actors'], inplace=True)

distances['start_actor_url'] = distances['start_actor'].apply(lambda x: actors[x])
distances['end_actor_url'] = distances['end_actor'].apply(lambda x: actors[x])

distances = distances[['start_actor', 'start_actor_url', 'end_actor', 'end_actor_url', 'distance']]

distances.to_csv('distances.csv')
```

## Full graph
---

```python
G_full = nx.Graph()

for (actor1, actor2), weight in actors_dist.items():
    G_full.add_edge(actor1, actor2, weight=weight)
weight_to_color = {
    1: 'green',
    2: 'blue',
    3: 'red',
    4: 'yellow',
    5: 'purple',
    6: 'orange',
    7: 'pink',
    8: 'brown',
    9: 'gray',
    10: 'cyan'
}

edge_colors = [weight_to_color.get(G_full[u][v]['weight'], 'black') for u, v in G_full.edges()]

pos = nx.spring_layout(G_full)  

nx.draw(G_full, pos, edge_color=edge_colors, width=5, 
        node_color='yellow', with_labels=True, node_size=2000)

edge_labels = nx.get_edge_attributes(G_full, 'weight')
nx.draw_networkx_edge_labels(G_full, pos, edge_labels=edge_labels)

plt.title('Actors Network Full Graph')
plt.savefig('graph_full.png')
plt.show()
```

### Collecting descriptions of movies by actors
---

```python
actors_descriptions = {}
for actor_name, actor_url in actors.items():
    actor_page_soup = dl_actor_page_soup(actor_url, log_handler=True)
    descriptions = get_movie_descriptions_by_actor_soup(actor_page_soup, actor_name, log_handler=True)
    actors_descriptions[actor_name] = descriptions
    time.sleep(30)
```

### Function for creating word_cloud
---

```python
stop_words = stopwords.words('english') + list(string.punctuation)
def generate_word_cloud(description, actor_name):
    #Preprocessing words
    words = word_tokenize(description)
    filtered_words = [word for word in words if word not in stop_words]
    text = ' '.join(filtered_words)

    #Creating word cloud
    wc = WordCloud(background_color='white', width=960, height=720)
    wordcloud = wc.generate(text)
    
    #Representation of wordcloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud)
    plt.axis('off')

    directory = 'actors_wordcloud'
    filename = actor_name + '.png'
    if not os.path.exists(directory):
        os.makedirs(directory)
    filepath = os.path.join(directory, filename)
    
    #Saving image
    plt.savefig(filepath)
    print(actor_name)
    plt.show()
```

### Generating images for each actor
---

```python
directory = 'actors_movie_description'
for actor_name in actors.keys():
    filename = actor_name + '.txt'
    filepath = os.path.join(directory, filename)
    with open(filepath, 'r') as f:
        description = f.read()
        generate_word_cloud(description, actor_name)
```
